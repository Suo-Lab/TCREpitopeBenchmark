```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import gridspec
import matplotlib.ticker as ticker

plt.rcParams['pdf.fonttype'] = 42

grouped_data = merged_df.groupby('Model')
fig, ax = plt.subplots(figsize=(7.3, 5))
colors = [
    '#CEBAF0', '#F38F3C','#A5405E',  '#68BF69', '#F7D1E3',
    '#BCD896', '#2279B0', '#D4242A', '#C0845C', '#734E9C',
    '#025939', '#592E13',  '#A9CCDE','#20B3AA', '#D2DC38']
markers = ['o', 's', 'D', '*', '^', 'v', '>', '<', 'p', 'h', 'x', 'P',  '1', '2','|']
for (name, group), color, marker in zip(grouped_data, colors, markers):
    if name == 'TCRGP':
        group = group[group['test_length'] <= 100000]  
    ax.plot(group['test_length'].astype(str), group['train_time(S)']/3600, label=name,
            marker=marker, linestyle='--', markersize=4, color=color, linewidth=1)
ax.set_title('Training Time for Different Models', fontsize=14)
ax.set_xlabel('Number of Training TCRs', fontsize=12)
ax.set_ylabel('Usage Time (log10 h)', fontsize=12) 
ax.set_yscale('log')
ax.yaxis.set_minor_locator(ticker.NullLocator())
ax.yaxis.set_minor_formatter(ticker.NullFormatter()) 
ax.yaxis.set_major_locator(ticker.LogLocator(base=10.0))
ax.tick_params(axis='x', labelsize=12)
ax.tick_params(axis='y', labelsize=12)
plt.legend(fontsize=10, loc='upper center', bbox_to_anchor=(0.5, -0.20), ncol=7)
plt.tight_layout()
path ="/home/luyanping/data/TCR_epitope_prediction/Compare_models_same_data/All_results/fig/fig6/"
plt.savefig(path + 'A_train_use_time.pdf', bbox_inches='tight')
plt.show()
```
